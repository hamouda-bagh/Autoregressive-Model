Question 1

```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
from scipy import signal

t = np.linspace(0, 0.1, 100)

f1 = np.cos(200 * m.pi * t + m.pi/4)
f2 = np.cos(200 * m.pi * t + m.pi/4) + 10 * np.cos(280 * m.pi * t)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(t,f1, 'r')

plt.plot(t,f2, 'b')

# show the plot
plt.show()
```

Question 2

```python
from scipy.fft import fft, fftfreq

fft_f1 = fft(f1,100)
fft_f2 = fft(f2,100)
print(len(fft_f2))
xf = fftfreq(100, 1/1000)
print(len(xf))

plt.plot(xf, 2.0/100 * np.abs(fft_f1))
plt.plot(xf, 2.0/100 * np.abs(fft_f2))
plt.grid()
plt.show()
```

```python
from scipy.fft import fft, fftfreq

fft_f1 = fft(f1,1024)
fft_f2 = fft(f2,1024)
print(len(fft_f2))
xf = fftfreq(1024, 1/1000)
print(len(xf))

plt.plot(xf, 2.0/1024 * np.abs(fft_f1))
plt.plot(xf, 2.0/1024 * np.abs(fft_f2))
plt.grid()
plt.show()
```

```python
fig,ax = plt.subplots(2)
t = np.linspace(-99/1000, 99/1000, 199)

corr1=np.correlate(f1,f1,'full')
corr2=np.correlate(f2,f2,'full')
ax[0].plot(t,corr1)
ax[1].plot(t,corr2)
ax[0].grid()
ax[1].grid()
plt.show()
```
